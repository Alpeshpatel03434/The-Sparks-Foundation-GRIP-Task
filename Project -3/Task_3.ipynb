### Split data Sub-Category with Sales and Profit

```python
data_group = dataset.groupby('Sub-Category')['Sales', 'Profit'].agg(sum)
data_group
```

**In above we aggregate all unique value and sum or adding each unique value of above data**

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle("Total Profit VS Sales ", fontsize=16)

sns.barplot(data = data_group, x = 'Sales', y ='Profit', ax=axes[1])
data_group.plot(kind='bar', ax = axes[0])

plt.xticks(rotation= 80, fontsize = 11)
plt.xlabel('Sales', fontsize = 12)
plt.ylabel('Profit', fontsize = 12)
plt.show
```

**We can see Profit vs Sales how co related with Sub-Category Features.**

### Split data Sub-Category with Sales and Quantity

```python
Sales_Quatity = dataset.groupby('Sub-Category')['Sales', 'Quantity'].agg(sum)
Sales_Quatity
```

**In above we aggregate all unique value and sum or adding each unique value of above data**

```python
fig,axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle("Total Sales VS Quantity", fontsize=16)

sns.barplot(data=Sales_Quatity, x='Sales', y='Quantity', ax=axes[1])
Sales_Quatity.plot(kind='bar',ax=axes[0])

plt.xticks(rotation= 80, fontsize = 11)
plt.xlabel('Sales', fontsize = 12)
plt.ylabel('Quantity', fontsize = 12)
plt.show()
```

**How to co related Quatity and Sales corealtion with sub-Category.**
