## Relationship between the data

```python
# correlation between variables
dataset.corr()
```

```python
# Histogram plot
dataset.hist(figsize=(50, 50))
plt.show()
```

```python
plt.figure(figsize=(16, 10))
sns.barplot(x = dataset['country_txt'].value_counts()[:20].index, y = dataset['country_txt'].value_counts()[:20].values)
plt.xticks(rotation = 60, fontsize = 12)
plt.title("Top Countries Got Attacked", fontsize=16)
plt.xlabel("Countries", fontsize = 14)
plt.ylabel("Number to Attacks", fontsize = 14)
plt.show()
```

In Above Chart The most number of Attacks done at Iraq Countrie   *and*      The less number of Attacks done at Egypt Countrie.

So We Can say that Egpt Countrie is more Sequre then Others

```python
plt.figure(figsize = (28, 20))
sns.countplot(x="country_txt", palette="coolwarm", data=dataset)
plt.xticks(rotation= 80)
plt.xlabel('country', fontsize = 14)
plt.ylabel('number of attacts', fontsize = 14)
plt.show()
```

```python
print("Terror Attacks From ", dataset.iyear.min(), "to", dataset.iyear.max())
```

```python
fig, axes = plt.subplots(1, 1, figsize=(20, 10))
sns.countplot(x="iyear", palette="coolwarm", data=dataset)
plt.xticks(rotation= 80)
plt.title("Terror Attacks From 1970 to 2017", fontsize=16)
plt.show()
```

In year 2014 the most number of Terror Attacks done

We can see that between 1970 to 1978 the Terror Attacks are less but after 1978 it's increasing number of Attacks

To 2012 to 2017 years between the Most Terror Attack done so We can say Security of countries is not good enough

## Death and Wound
